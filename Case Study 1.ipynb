```python
#16.	Find:
##a.	Mean and std deviation of the income
##b.	For income greater than or equal to 10.5, find the mean income
##c.	For each brand having price >=2.5 find the mean, median, sd of the log of income

#a.
print("Mean:\n",oj.INCOME.mean())
print("STD:\n",oj.INCOME.std())

#b.
print("Mean_of_Income>10.5:\n",oj[oj.INCOME>10.5].INCOME.mean())

#c
print("Mean_LogInc:\n",oj[oj.price>=2.5].LogInc.mean())
print("Median_LogInc:\n",oj[oj.price>=2.5].LogInc.median())
print("StdDev_LogInc:\n",oj[oj.price>=2.5].LogInc.std())
```
