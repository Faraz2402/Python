```python
#Q.4)	In a single terror incident in India, find out top 5 cities by number killed

t4 = terror.query("(country_txt=='India')")[["eventid","country_txt","city","nkill"]].sort_values('nkill',ascending=False)
#t5['nkill_nwound']=t5.nkill+t5.nwound
t4.sort_values('nkill',ascending=False).head(5)
```

```python
#Q.5)	In a single terror incident in India, find out top 5 cities by number killed and wounded 

t5=terror.query("(country_txt=='India')")[["eventid","country_txt","city","nkill","nwound"]].sort_values('nkill',ascending=False)
t5['nkill_nwound']=t5.nkill+t5.nwound
t5.sort_values('nkill_nwound',ascending=False).head(5)
```

```python
#Q.8) Write a function to label an incident that was both successful and suicidal


def check(event) :
    if ((terror['suicide'][terror['eventid'] == event] == 1).tolist()[0] and 
        (terror['success'][terror['eventid'] == event] == 1).tolist()[0]) :
        print('Yes')
    else :
        print('No')
        
check(201410290055)
```
